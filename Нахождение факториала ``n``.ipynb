Факториал – это произведение всех целых чисел от 1 до данного числа(данное число здесь это «n»).

Факториала формула.jpg

```python
factorial_of = 3 # наше число "n", факториал которого мы будем находить
```

```python
factorial = 1 # задаём старт
for i in range(2, factorial_of + 1):
  factorial *= i

print(factorial) # выводим сам факториал
```
