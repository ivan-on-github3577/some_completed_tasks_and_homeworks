```python
from random import randint
```

```python
# Для начала создадим список случайных чисел.
numbers = []
for i in range(1000):
  numbers.append(randint(1, 10000))

print(numbers)
```

```python
largest_number = 0
almost_largest_number = 0
for index in range(len(numbers)):
  if numbers[index] > largest_number:
    largest_number = numbers[index]
  if numbers[index] > almost_largest_number:
    if numbers[index] < largest_number:
      almost_largest_number = numbers[index]

print('Первое по величине число:', largest_number)
print('ВТОРОЕ ПО ВЕЛИЧИНЕ ЧИСЛО:', almost_largest_number)
```
